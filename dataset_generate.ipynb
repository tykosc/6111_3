```python
s = pd.read_csv('./datasets/squirrel.csv')
s = s[['Shift', 'Age', 'Primary Fur Color', 'Highlight Fur Color', 'Location', 'Running', 'Chasing', 'Climbing', 'Eating', 'Foraging', 'Kuks', 'Quaas', 'Moans', 'Tail flags', 'Tail twitches', 'Approaches', 'Indifferent', 'Runs from']]
#s['Running'] = s['Running'].replace({True: 'Yes', False: 'No'})
s
```

```python
for col in ['Running', 'Chasing', 'Climbing', 'Eating', 'Foraging', 'Kuks', 'Quaas', 'Moans', 'Tail flags', 'Tail twitches', 'Approaches', 'Indifferent', 'Runs from']:
    s[col] = s[col].replace({True: col, False: ''})

    #s[col] = s[col].replace([True], col, inplace=True)
    #s[col] = s[col].map({True: 'True', False: 'False'})
s
```

```python
s.replace({None: ''})
```

```python
s.to_csv(path_or_buf='./datasets/sqr2.csv', header=False, index=False)
```
